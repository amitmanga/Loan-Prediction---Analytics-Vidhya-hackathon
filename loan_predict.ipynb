# Loan Prediction

```python
#importing numpy and pandas libraries

import pandas as pd
import numpy as np
```

```python
#importing required dataset using pandas

train = pd.read_csv('loan_dataset.csv.csv')
```

```python
train.head(5)
```

```python
#From Sklearn library importing labelEncoder to convert categorical string columns into numeric
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['Education'] = le.fit_transform(train['Education'].astype('str'))
```

```python
train.head()
```

```python
train['Education'] = le.fit_transform(train['Education'].astype('str'))
train['Gender'] = le.fit_transform(train['Gender'].astype('str'))
train['Self_Employed'] = le.fit_transform(train['Education'].astype('str'))
train['Married'] = le.fit_transform(train['Married'].astype('str'))
train['Property_Area'] = le.fit_transform(train['Property_Area'].astype('str'))
```

```python
train.head(5)
```

```python
#Dropping out few columns of dataset
col=['Gender','Married','Education','Self_Employed','Property_Area','Loan_ID','Dependents']
train_new= train.drop(col,axis=1)
train_new.head()
```

```python
#Filling NaN values by Zero
train_new = train_new.fillna(0)
train_new.head()
```

```python
import xgboost as xgb
from sklearn.model_selection import  train_test_split
from xgboost.sklearn import  XGBClassifier
```

```python
#Assigning target variable into Y
X=train_new.iloc[:,0:5] 
Y=train_new[['Loan_Status']]
X.head()
```

```python
#Splitting the dataset into train and test
from sklearn.model_selection import  train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)
X_train.head()
```

```python
#Importing xgboost Classifier 
from xgboost.sklearn import XGBClassifier
gbm = XGBClassifier().fit(X_train,Y_train)
predictions = gbm.predict(X_train)
print(predictions)
```

```python
#Checking score on test dataset
gbm.score(X_test, Y_test)
```

```python
#using RandomForest classifier to check accuracy
from sklearn.ensemble import RandomForestClassifier as RF
rf = RF().fit(X_train,Y_train)
prediction = rf.predict(X_train)
```

```python
#Checking score on test dataset.
rf.score(X_test, Y_test)
```
